# Prediction with KNN

```python
haberman.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
neigh = KNeighborsClassifier()
```

```python
X = haberman[['Age', 'Year of Operation', 'No of axillary Nodes']]
Y = haberman['Survival Years']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
neigh.fit(X_train, Y_train)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_test = scaler.fit_transform(X_test)
X_train = scaler.transform(X_train)
```

```python
X_train.shape
```

```python
Y_Prediction = neigh.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(Y_Prediction, Y_test)
print(acc)
```

# Prediction with KNN Parameter

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

accuracies_test = []
accuracies_train = []

for i in range(1,25):
    neigh = KNeighborsClassifier(n_neighbors=i)
    neigh.fit(X_train, Y_train)
    
    Y_Pred_test = neigh.predict(X_test)
    acc_test = accuracy_score(Y_Pred_test, Y_test)
    
    Y_Pred_train = neigh.predict(X_train)
    acc_train = accuracy_score(Y_Pred_train, Y_train)
    
    accuracies_test.append(acc_test)
    print('for test', i, 'accuracy is ',acc_test)
    
    accuracies_train.append(acc_test)
    print('for train', i, 'accuracy is ',acc_train)
    print('*'*100)
    
accuracy_index = accuracies_test.index(max(accuracies))+1
print('max score is ', max(accuracies_test),'and i value is', accuracy_index)
```

```python
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X_train, Y_train)
Y_Prediction = neigh.predict(X_test)
from sklearn.metrics import accuracy_score
acc = accuracy_score(Y_Prediction, Y_test)
print(acc)
```
